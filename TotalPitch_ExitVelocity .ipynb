```python
import requests
import json


request=requests.get('https://statsapi.mlb.com/api/v1.1/game/633709/feed/live')
data_json = request.json()
data_str = json.dumps(data_json, indent=2)
print(data_str)
#print(data_json)
```

```python
from pandas.io.json import json_normalize

live_data = data_json['liveData']['plays']['allPlays']
live_data_str = json.dumps(live_data[0], indent=2)
playevent = live_data[0]['playEvents']
playevent_str = json.dumps(playevent, indent=2)
print(playevent_str)
```

```python
import pandas as pd

df_playevent = pd.DataFrame(json_normalize(playevent))
#print(df_playevent)
```

```python
import requests
import json

response = requests.get('https://statsapi.mlb.com/api/v1.1/game/633709/feed/live')

data = response.json()

launchSpeed = []
pitchNumber = []
for instance in data['liveData']['plays']['allPlays']:
    for inner_instance in instance['playEvents']:
        try:
            launchSpeed.append(inner_instance['hitData']['launchSpeed'])
            pitchNumber.append(inner_instance['pitchNumber'])
        except:
            pass
print(launchSpeed)
print(pitchNumber)
```

```python
highest_exit_velocity = max(launchSpeed)
total_pitch = sum(pitchNumber)
print(highest_exit_velocity)
print(total_pitch)
```
